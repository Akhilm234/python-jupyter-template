```python
#activity 1

### Exercise 1: Line Plot Customization
import matplotlib.pyplot as plt

# Create x values from 0 to 10
x = list(range(0, 11))

# Calculate y values as the square of x
y = [i**2 for i in x]

# Create the line plot
plt.plot(x, y)

# Add title and labels
plt.title('Line Plot of X vs X^2')
plt.xlabel('X')
plt.ylabel('Y = X^2')

# Add grid
plt.grid(True)

# Display the plot
plt.show()
# TODO: Create a line plot with x values ranging from 0 to 10 and y values as the square of x.
# Customize the plot by adding a title, labels for both axes, and a grid.
```

```python
# activity 2
### Exercise 2: Bar Plot with Colors
# TODO: Create a bar plot with the following data: categories = ['A', 'B', 'C', 'D'] and values = [5, 7, 3, 9].
# Use different colors for each bar and add a title to the plot.
import matplotlib.pyplot as plt

# Define categories and values
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]

# Create the bar plot with different colors
plt.bar(categories, values, color=['red', 'blue', 'green', 'orange'])

# Add title to the plot
plt.title('Bar Plot of Categories and Values')

# Display the plot
plt.show()
```

```python
#activity 3
### Exercise 3: Scatter Plot with Annotations
# TODO: Create a scatter plot with x = [1, 2, 3, 4, 5] and y = [2, 4, 6, 8, 10].
# Annotate each point with its (x, y) value, and set the title as 'Scatter Plot with Annotations'.
import matplotlib.pyplot as plt

# Define x and y values
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Create the scatter plot
plt.scatter(x, y)

# Annotate each point with its (x, y) value
for i, j in zip(x, y):
    plt.annotate(f'({i}, {j})', (i, j), textcoords="offset points", xytext=(5, 5), ha='center')

# Set title
plt.title('Scatter Plot with Annotations')

# Display the plot
plt.show()
```

```python
#activity 4
### Exercise 4: Pie Chart with Percentages
# TODO: Create a pie chart with the following data: labels = ['Python', 'Java', 'C++', 'JavaScript'] and sizes = [40, 25, 20, 15].
# Display the percentages on the chart using autopct.
import matplotlib.pyplot as plt

# Define labels and sizes
labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]

# Create pie chart with percentages displayed
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)

# Add title
plt.title('Pie Chart of Programming Languages')

# Ensure the circle's proportion
plt.axis('equal')

# Display the chart
plt.show()
```

```python
#activity 5
### Exercise 5: Subplot Layout
# TODO: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
import matplotlib.pyplot as plt
import numpy as np

# Sample data for plotting
x = np.arange(1, 6)
y = np.random.randint(1, 10, size=5)  # Random data for line and bar charts
labels = ['A', 'B', 'C', 'D', 'E']  # Update labels to match the length of y

# Create a 2x2 subplot layout
fig, axs = plt.subplots(2, 2, figsize=(10, 10))

# First subplot: Line chart
axs[0, 0].plot(x, y, marker='o')
axs[0, 0].set_title('Line Chart')
axs[0, 0].set_xlabel('X-axis')
axs[0, 0].set_ylabel('Y-axis')

# Second subplot: Bar chart
axs[0, 1].bar(x, y)
axs[0, 1].set_title('Bar Chart')
axs[0, 1].set_xlabel('X-axis')
axs[0, 1].set_ylabel('Y-axis')

# Third subplot: Scatter plot
axs[1, 0].scatter(x, y, color='r')
axs[1, 0].set_title('Scatter Plot')
axs[1, 0].set_xlabel('X-axis')
axs[1, 0].set_ylabel('Y-axis')

# Fourth subplot: Pie chart
# Use the first 5 values of y for the pie chart and update labels accordingly
axs[1, 1].pie(y, labels=labels, autopct='%1.1f%%')
axs[1, 1].set_title('Pie Chart')

# Adjust layout and show the plot
plt.tight_layout()
plt.show()
```
