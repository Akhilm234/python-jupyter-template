### Exercise 1: Line Plot Customization

```python
import matplotlib.pyplot as plt

# Create x values from 0 to 10
x = list(range(0, 11))

# Calculate y values as the square of x
y = [i**2 for i in x]

# Create the line plot
plt.plot(x, y)

# Add title and labels
plt.title('Line Plot of X vs X^2')
plt.xlabel('X')
plt.ylabel('Y = X^2')

# Add grid
plt.grid(True)

# Display the plot
plt.show()
```
